```python
import matplotlib.pyplot as plt
import numpy as np
# Define the functions
def f1(x):
    return x*200

def f2(x):
    return 1650+((x-1)*55)

# Generate x values
x = np.linspace(0, 20, 100)

# Generate y values for f1(x)
y1 = f1(x)

# Generate y values for f2(x)
y2 = f2(x)

# Plot the functions
plt.plot(x, y1, label='f(x) = x^2')
plt.plot(x, y2, label='f(x) = x + 1')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) = x^2 and f(x) = x + 1')
plt.legend()


# Save the plot
plt.savefig('plot.png')
```
